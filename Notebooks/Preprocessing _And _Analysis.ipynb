```python
unique_values_data= pd.DataFrame({
    'Column': data.columns,
    'Unique Values': [data[col].unique() for col in data.columns],
    'Count': [data[col].nunique() for col in data.columns]
})
unique_values_data
```

```python
duplicates = data.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates}")
```

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

outliers = {}

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers[col] = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    print(f"Column: {col}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Number of outliers: {outliers[col].shape[0]}\n")
```

#  Preprocessing and Feature Engineering:

```python
data['SpendingPerMonth'] = data['TotalCharges'] / (data['tenure'] + 1)  
```

```python
cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in cols:
    data[col] = data[col].replace('No internet service', 'No')
```

```python
cols = ['MultipleLines']

for col in cols:
    data[col] = data[col].replace('No phone service', 'No')
```

```python
cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn', 'gender',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV',
        'StreamingMovies','MultipleLines']
for col in cols:
    data[col] = data[col].map({'Yes': 1, 'No': 0, 'Female': 0, 'Male': 1,'True': 1 ,'False': 0})

# unique_values_data = pd.DataFrame({
#     'Column': data.columns,
#     'Unique Values': [list(data[col].unique()) for col in data.columns],
#     'Count': [data[col].nunique() for col in data.columns]
# })
```

```python
data.to_csv("cleaned data before encoding and scalling.csv", index=False) 
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['InternetService', 'Contract']
encoder = LabelEncoder()
for col in label_cols:
    data[col] = encoder.fit_transform(data[col])
```

One-Hot Encoding 

```python
data = pd.get_dummies(data, columns=['PaymentMethod'], drop_first=True)
```

```python
unique_values_data = pd.DataFrame({
    'Column': data.columns,
    'Unique Values': [list(data[col].unique()) for col in data.columns],
    'Count': [data[col].nunique() for col in data.columns]
})

unique_values_data
```

Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[['MonthlyCharges', 'TotalCharges', 'tenure','SpendingPerMonth']] = scaler.fit_transform(data[['MonthlyCharges', 'TotalCharges','tenure','SpendingPerMonth']])
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
```

```python
# Load the cleaned dataset
data = pd.read_csv("cleaned data after encoding and scalling.csv")

# Display basic information about the dataset
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows of the Dataset:")
print(data.head())
```

```python
# Dataset Overview
print("Dataset Shape:", data.shape)
print("\nData Types:\n", data.dtypes)
print("\nMissing Values:\n", data.isnull().sum())
```

```python
# Descriptive Statistics
print("\nDescriptive Statistics for Numerical Columns:\n", data.describe())
```

```python
# Target Variable Distribution
sns.countplot(x='Churn', data=data)
plt.title("Distribution of Churn")
plt.show()

# Print value counts
print("\nChurn Value Counts:")
print(data['Churn'].value_counts())
```

```python
# Drop non-numeric columns (e.g., customerID) and select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Histograms for Numerical Features
data.hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# Boxplots for Numerical Features vs Churn
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    if column != 'Churn':
        plt.figure(figsize=(8, 4))
        sns.boxplot(x='Churn', y=column, data=data)
        plt.title(f"{column} vs Churn")
        plt.show()
```
